Discuss string slicing and provide examples? String slicing in Python allows you to extract parts of a string by specifying a start index, an end index, and an optional step. The general syntax for slicing is:
python string[start:end:step]

start: The index where the slice begins (inclusive).
end: The index where the slice ends (exclusive).
step: The interval between indices (optional).
Here are some examples:

Basic Slicing:

python text = "Hello, World!" slice1 = text[0:5] # Extracts "Hello"

In this example, text[0:5] extracts the substring starting from index 0 up to but not including index 5.

Omitting Start or End:

python slice2 = text[:5] # Extracts "Hello" (same as text[0:5]) slice3 = text[7:] # Extracts "World!" (from index 7 to the end)

If you omit the start index, it defaults to 0. If you omit the end index, it defaults to the end of the string.

Using Step:

python slice4 = text[::2] # Extracts "Hlo ol!" (every second character)

Here, text[::2] extracts characters starting from the beginning to the end, but only every second character.

Negative Indices and Slicing:

python slice5 = text[-6:-1] # Extracts "World" (negative indices count from the end)

Negative indices count from the end of the string. -1 refers to the last character.

These examples demonstrate how string slicing provides a flexible way to manipulate and access specific parts of a string in Python.

Explain the key features of lists in python? Lists in Python are versatile and powerful data structures. Here are some key features of lists:

Ordered: Lists maintain the order of elements. The position of each element is preserved as you add, remove, or modify elements.

python my_list = [1, 2, 3, 4]

Mutable: You can modify the elements of a list after its creation. This means you can change, add, or remove elements.

python my_list[1] = 5 # Changes the second element to 5 my_list.append(6) # Adds 6 to the end of the list my_list.remove(3) # Removes the element 3 from the list

Dynamic Size: Lists can grow or shrink in size as elements are added or removed. There's no need to specify the size when creating a list.

python dynamic_list = [] dynamic_list.append(1) dynamic_list.append(2)

Heterogeneous Elements: Lists can store elements of different data types. You can mix integers, strings, and other types in a single list.

python mixed_list = [1, "hello", 3.14, [2, 3]]

Indexing and Slicing: Lists support indexing and slicing to access or modify specific elements or sublists.

python first_element = my_list[0] # Accesses the first element sublist = my_list[1:3] # Creates a sublist from index 1 to 2

Nested Lists: Lists can contain other lists, allowing for the creation of multi-dimensional structures.

python nested_list = [[1, 2], [3, 4], [5, 6]]

Methods: Lists come with a variety of built-in methods for common operations like adding, removing, and sorting elements.

python my_list.sort() # Sorts the list in ascending order my_list.reverse() # Reverses the order of elements

These features make lists a fundamental and highly flexible data structure in Python.

Discuss the importants of dictionary keys being immutable and provide examples. In Python, dictionary keys must be immutable because mutable objects, such as lists, can change their value and thereby affect the integrity of the dictionary’s structure. Here’s why immutability is crucial and some examples to illustrate:
Importance of Immutable Keys
Hash Stability: Dictionary keys must be immutable to ensure that their hash value remains constant. The dictionary uses these hash values to quickly locate and retrieve values associated with keys. If a key were mutable and could change, its hash value would also change, leading to potential inconsistencies and errors in accessing the data.

Data Integrity: Immutable keys guarantee that the association between a key and its value remains consistent throughout the lifetime of the dictionary. This ensures that the dictionary functions correctly and reliably.

Examples
Immutable Keys:

python my_dict = {('name', 'last'): 'Alice', ('name', 'first'): 'Bob'}

Here, tuples are used as keys. Since tuples are immutable, they can be reliably used as dictionary keys.

Mutable Keys (Not Allowed):

python my_dict = {[1, 2]: 'value'} # This will raise a TypeError

Lists are mutable and cannot be used as dictionary keys. If you try to use them, Python will raise a TypeError.

String as a Key:

python my_dict = {'name': 'Alice', 'age': 30}

Strings are immutable, so they are valid dictionary keys. The dictionary maintains the correct mapping of these immutable keys to their corresponding values.

Custom Immutable Key Example:

python class ImmutableKey: def init(self, value): self.value = value def hash(self): return hash(self.value) def eq(self, other): return isinstance(other, ImmutableKey) and self.value == other.value my_key = ImmutableKey('unique_key') my_dict = {my_key: 'some_value'}

In this example, a custom class ImmutableKey is used as a dictionary key. The hash and eq methods ensure that instances of this class are immutable and can be used as dictionary keys.

Summary
Immutable Keys: Ensures stable hash values and reliable data access.
Mutable Keys: Cannot be used because they can change and disrupt the dictionary's functionality.
Using immutable keys guarantees that the dictionary’s internal structure remains consistent and functional.
